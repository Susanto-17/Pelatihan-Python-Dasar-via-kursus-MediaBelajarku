# Pandas Library

```python
import pandas as pd
```

```python
# Read csv file
df = pd.read_csv("titanic.csv")
df.head()
```

# Filtering Baris

```python
# Filtering by Pclass == 1
pclass1 = (df.Pclass == 1) 
```

```python
data_pclass1 = df[pclass1]
# print(data_pclass1[:10])
print(len(data_pclass1))
```

```python
# Filtering by ages menggunakan isin operator
ages = [18, 21, 28]

# ages = [i for i in range(10,21)]

data_age = df[df["Age"].isin(ages)]
data_age.head()
print(len(data_age))
```

```python
data_age.count()
```

```python
gender = "male"
data_age_pclass1 = df[df["Age"].isin(ages) & pclass1 & (gender == 'male')]
data_age_pclass1.head()
print(len(data_age_pclass1))
```

```python
data_age_pclass1.count()
```

# Filtering Kolom

```python
df_filter_kolom = df.filter(["Name", "Sex", "Age"])
df_filter_kolom.head()
```

# Menggabungkan Dataframes

```python
titanic_pclass1 = df_filter_kolom[df.Pclass == 1]
print(titanic_pclass1.shape)

titanic_pclass2 = df_filter_kolom[df.Pclass == 2]
print(titanic_pclass2.shape)
```

```python
# Menggunakan append method
gabung_pclass12 = titanic_pclass1.append(titanic_pclass2, ignore_index=True)
print(gabung_pclass12)
```

```python
# Menggunakan concat method
gabung2_pclass12 = pd.concat([titanic_pclass1, titanic_pclass2])
print(gabung2_pclass12)
```

# Sorting / Mengurutkan Dataframes

```python
# Menggunakan sort_values method
data_sort_by_age = df.sort_values(by=["Age"])
data_sort_by_age.head()
```

```python
data_sort_by_age = df.sort_values(by=["Age"], ascending = False)
data_sort_by_age[:5]
```

# Pivot dan Crosstab

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_pivot = df.pivot_table(index='Sex', columns='Pclass', values='Fare')
print(data_pivot)
```

```python
flight_data = sns.load_dataset('flights')
print(flight_data)
```

```python
data_flight_pivot = flight_data.pivot_table(index='month', columns='year', values='passengers')
print(data_flight_pivot)
```
